```python
import json

topics_file = open("machine_learning_topics.json")
data = json.load(topics_file)

topic_ids = [10057, 10538, 10682, 10824, 11689, 12072, 12535, 12676, 13373, 13702]
```

```python
print(f"{"=" * 40}\nTopics I chose\n{"=" * 40}")
for topic in data:
    id = topic['id'].split("/T")[1]
    if int(id) in topic_ids:
        topic_str = f"{id}) {topic['display_name']}"
        works_str = f"{topic['works_count']} works"
        cited_by_str = f"{topic['cited_by_count']} cited_by"
        print(f"{topic_str:<55} {works_str:<15} {cited_by_str:<20}")
```

```python
ids = []
topic_works_counts = []
topic_cited_by_counts = []
for topic in data:
    id = topic['id'].split("/T")[1]
    works = (f"{id}) {topic['display_name']}", topic['works_count'])
    topic_works_counts.append(works)
    cited_by = (f"{id}) {topic['display_name']}", topic['cited_by_count'])
    topic_cited_by_counts.append(cited_by)

sorted_by_works = sorted(topic_works_counts, key=lambda x: x[1], reverse=True)
sorted_by_cited = sorted(topic_cited_by_counts, key=lambda x: x[1], reverse=True)
```

```python
print(f"{"=" * 40}\nTop topics by cited_by\n{"=" * 40}")
for i in range(20):
    print(f"{sorted_by_cited[i][0]} = {sorted_by_cited[i][1]}")
print(f"\n{"=" * 40}\nTop topics by num works\n{"=" * 40}")
for i in range(20):
    print(f"{sorted_by_works[i][0]} = {sorted_by_works[i][1]}")
```
